###  Numpy + Matplotlib

```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# exp
x = np.arange(-2,2,0.01)
plt.title('y=a**x')
for a in range(1,4):
    y = np.power(a,x)
    plt.plot(x,y,label='y={0}**x'.format(a))
plt.grid()
plt.legend(loc='upper center')
plt.show()

# log
x = np.arange(0.1,5,0.01)
plt.title('y=log(x)')
for a in range(1,4):
    y = a * np.log(x)
    plt.plot(x,y,label='y={0}*log(x)'.format(a))
    
plt.grid()
plt.legend(loc='lower center')
plt.show()
```
